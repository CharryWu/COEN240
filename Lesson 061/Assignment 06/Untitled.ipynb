```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([
    [1,1,2],
    [1,2,4],
    [1,-3,-6]
])
```

```python
XTX = X.T @ X
XTX
```

```python
y = np.array([[1],[2],[-3]])
```

```python
XTy = X.T@y
XTy
```

```python
I = np.identity(3)
```

```python
lambs = [10**-5, -10**-5, 10**-4, -10**-4, 10**-3, -10**-3, 10**-2, -10**-2, 10**-1, -10**-1]
```

```python
betas = []
Es = []
```

```python
for lamb in lambs:
    beta = np.linalg.inv(XTX + lamb*I) @ (XTy)
    E = (y-X @ beta).T @ (y-X @ beta)
    betas.append(beta)
    Es.append(E[0][0])
    print('lambda=',lamb)
    print(beta)
    print('E=',E[0][0])
    print()
```

```python

# convert y-axis to Logarithmic scale
plt.yscale("log")
plt.xscale("log")
 
plt.scatter(Es, lambs)
```

```python
# convert y-axis to Logarithmic scale
plt.yscale("log")
plt.xscale("log")
 
plt.scatter([Es[i] for i in range(len(Es)) if i % 2 == 0], [lambs[i] for i in range(len(lambs)) if i % 2 == 0])
```

```python
# convert y-axis to Logarithmic scale
plt.yscale("log")
plt.xscale("log")
 
plt.scatter([Es[i] for i in range(len(Es)) if i % 2 == 1], [-lambs[i] for i in range(len(lambs)) if i % 2 == 1])
```
