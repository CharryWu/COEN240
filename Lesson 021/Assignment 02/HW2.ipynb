```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
excel_data_df = pd.read_excel('Data Points.xlsx', sheet_name='Sheet1')
```

```python
excel_data_df
```

```python
excel_data_df.columns
```

```python
df = excel_data_df[['x-cord', 'y-cord']]
df
```

```python
df_np = df.to_numpy()
df_np
```

```python
# plot
plt.scatter(
   df_np[:, 0], df_np[:, 1],
   c='white', marker='o',
   edgecolor='black', s=50
)
plt.show()
```

```python
# Perform K-means clustering
kmeans = KMeans(n_clusters=3)
res = kmeans.fit_predict(df)
res
```

```python
df_labels = pd.DataFrame(res, columns = ['predict_label'])
result = pd.concat([excel_data_df[['Label']], df, df_labels], axis=1)
result
```

```python
kmeans.cluster_centers_
```

```python
num_iterations = kmeans.n_iter_
num_iterations
```

```python
cluster0 = result.loc[result['predict_label'] == 0]
cluster0
```

```python
kmeans.cluster_centers_[0]
```

```python
cluster1 = result.loc[result['predict_label'] == 1]
cluster1
```

```python
kmeans.cluster_centers_[1]
```

```python
plt.scatter(
    df_np[res == 0, 0], df_np[res == 0, 1],
    s=50, c='lightgreen',
    marker='s', edgecolor='black',
    label='cluster 1'
)

plt.scatter(
    df_np[res == 1, 0], df_np[res == 1, 1],
    s=50, c='orange',
    marker='o', edgecolor='black',
    label='cluster 2'
)

plt.scatter(
    df_np[res == 2, 0], df_np[res == 2, 1],
    s=50, c='lightblue',
    marker='v', edgecolor='black',
    label='cluster 3'
)

# plot the centroids
plt.scatter(
    kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
    s=250, marker='*',
    c='red', edgecolor='black',
    label='centroids'
)

plt.show()
```
